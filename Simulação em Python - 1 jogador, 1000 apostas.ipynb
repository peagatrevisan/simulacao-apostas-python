```python
#EV = (ganho × chance de ganhar) – (perda × chance de perder)
```

```python
#EV = (20 * 0.3) - (10 * 0.7) = 6 – 7 = -1
```

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
valor_aposta = 10
valor_premio = 20
chance_ganhar = 0.3
n_apostas = 1000
```

```python
saldo = 0
historico = []
for _ in range(n_apostas):
    if np.random.rand() < chance_ganhar:
        saldo += valor_premio
    else:
        saldo -= valor_aposta
    historico.append(saldo)
```

```python
plt.figure(figsize=(12, 6.75))
plt.plot(historico, color='blue', linewidth=2)
plt.axhline(0, color='black', linestyle='--')
plt.title("Imagem 1 - Saldo Acumulado: 1 Jogador, 1000 Apostas")
plt.xlabel("Número de Apostas")
plt.ylabel("Saldo (R$)")
plt.grid(True)
plt.tight_layout()
plt.show()
```
